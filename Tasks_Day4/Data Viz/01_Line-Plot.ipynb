# Adding Limits

```python
#import librasries 
import seaborn as sns
import matplotlib.pyplot as plt

#load data set
phool=sns.load_dataset("iris")
phool

#draw line graph
sns.lineplot(x="petal_length",y="petal_width",data=phool)
#title adding
plt.title("phoolo ka plot")

#adding limit
plt.xlim(2)
plt.ylim(1)
#show plot
plt.show()
```

```python
#import librasries 
import seaborn as sns
import matplotlib.pyplot as plt

#load data set
phool=sns.load_dataset("iris")
phool

#draw line graph
sns.lineplot(x="petal_length",y="petal_width",data=phool)
#title adding
plt.title("phoolo ka plot")
#style set

sns.set_style("whitegrid")
plt.show()

#adding limit
plt.xlim(2)
plt.ylim(1)

#show plot
plt.show()
```

# size of fig

```python
#import librasries 
import seaborn as sns
import matplotlib.pyplot as plt

#load data set
phool=sns.load_dataset("iris")
phool
# set fig size
plt.figure(figsize=(4,7))
#draw line graph
sns.lineplot(x="petal_length",y="petal_width",data=phool)
#title adding
plt.title("phoolo ka plot")

#adding limit
plt.xlim(2)
plt.ylim(1)
#show plot
plt.show()
```
