```python
ks1=kashti[["Age in Days","fare"]]
ks1.head()
```

## Method of Normalization
1. Simple feature scaling
    - x(new)=x(old)\x(max)
2. Min-Max Method
3. Z- score standards
4. log transformation

## Some other Methods of Normalization Data:

1- By taking square root of data or Log of Data

2- Feature Clipping

3- Removing Outliers

```python
# Simple feature scaling
ks1["fare"]=ks1["fare"]/ks1["fare"].max()
ks1["Age in Days"]=ks1["Age in Days"]/ks1["Age in Days"].max()
ks1.head()
```

```python
# Min -MAx MEthod
ks1["fare"]=(ks1["fare"]-ks1["fare"].min())/(ks1["fare"].max()-ks1["fare"].min())
ks1.head()
```

```python
#Z- Score
ks1["fare"]=(ks1["fare"]-ks1["fare"].mean())/ks1["fare"].std()
ks1.head()
```

```python
# Log transmation
ks1["fare"]=np.log(ks1["fare"])
ks1.head()
```

# Binning
- Grouping the values into smaller number of values(bins)
- covert numeric into catogorical
- to have better understanding of groups
    - low vs mid vs high

```python
# Assingment baba g
bins=np.linspace(min(kashti2["age"]),max(kashti2["age"]),4)
age_groups=["bachy","jawan","boorhy"]
kashti2["age"]=pd.cut(kashti2["age"],bins,labels=age_groups,include_lowest=True)
kashti2["age"]
```
